```python
from gurobipy import *
import pandas as pd
import numpy as np
import math
global epsilon
global m
global size1
global r
epsilon=0.000000001
instances=1
sheet = pd.DataFrame(columns=['Instance', 'Tightness', 'Alpha','Makespan'])
```

```python
units=[2]
nodes=['one', 'two', 'three', 'four', 'five', 'six']; 
arcs=tuplelist([('one','two'),('one' ,'three'),('two', 'one'),('two','four'),('two','five'),
      ('three', 'one'),( 'three', 'four'),('three', 'five' ),('four','two'),
      ('four','three'),('four','six'),('five','two'),('five','three'),('five','six'),('six','four'),('six','five')])
c={}
for arc in arcs:
    c[arc]=10
    
NoFlows=5
orig=['four','six','one','four','three']
dest =['one','five','five','two','four']
size=np.array([66,58,94,36,1])
VV=[20.000000, 20.000000, 20.000000,30.000000,30.000000]
deadline=np.array([3.300000,2.900000,4.700000,1.200000,0.033333])
tightness=[0,30]
MinAlpha=1.684570
lowerbound={0:5.969,30:5.4}
ExistPathFlag=0;
size1=np.zeros(NoFlows)
```

```python
#creating model: 
m = Model('netflow')
m.setParam('OutputFlag', 0) 
# Create variables
r = m.addVars(NoFlows, name="r",lb=0,ub=float('inf'))
y = m.addVars(NoFlows,arcs, name="y",lb=0)
z = m.addVars(NoFlows,arcs,units, name="z",lb=0,vtype=GRB.INTEGER)
#create constraints
capacity1=m.addConstrs(
    (sum(u*z[f,i,j,u] for u in units)>=y[f,i,j] for f in range(NoFlows) for i,j in arcs ), "capacity1")

capacity2=m.addConstrs(
    (sum(u*z[f,i,j,u] for u in units for f in range(NoFlows))<=c[i,j] for i,j in arcs ), "capacity2")

balance_orig=m.addConstrs(
    (y.sum(f,'*',j)-y.sum(f,j,'*')==-r[f] for f in range(NoFlows) for j in [orig[f]] ), "balance_orig")
balance_dest=m.addConstrs(
    (y.sum(f,'*',j)-y.sum(f,j,'*')==r[f] for f in range(NoFlows) for j in [dest[f]] ), "balance_dest")

balance_other=m.addConstrs(
    (y.sum(f,'*',j)-y.sum(f,j,'*')==0 
     for f in range(NoFlows) for j in list(set(nodes)-set().union([orig[f]],[dest[f]])) ), "balance_other")
print('model is created.')
```

```python
for ins  in range(instances):
    for t in tightness:
        print('\n')
        deadline1=deadline.copy()
        alpha=(MinAlpha+t*MinAlpha/100)
        deadline1=alpha*deadline
        makespan=0
        feasible=0
        size1=size.copy()
        print('size=',size)
        print('original deadline=',deadline)
        print('deadline=',deadline1)
        print('makespan:%8.3f'%makespan)

        for iter in range(NoFlows):
            print('-------------------------------------------------------')
            print('iter:%3d , tightness:%8.3f , alpha:%8.3f'%(iter,t,alpha))
            print('-------------------------------------------------------')
            print('solve for size:', size1)
            m.update()
            #objective function
            m.setObjective(sum((1/deadline1[f]**2)*r[f] for f in range(NoFlows)), GRB.MAXIMIZE)
            for f in range(NoFlows):
                if size1[f]<0.0001:
                    r[f].ub=0
                else:
                    r[f].ub=float('inf')
            m.optimize()
            print(m.objVal)
            m.write('netflow.lp')
            print('model solved...')
            time=float('inf')*np.ones(NoFlows)
            rate=np.zeros(NoFlows)
            for f in range(NoFlows):
                if r[f].x>0:
                    rate[f]=r[f].x
                    time[f]=size1[f]/r[f].x
            print('rate=',rate)
            print('time=',time)
            print('min_time:%8.3f in index:%8.3f with deadline:%8.3f'%(min(time) ,np.argmin(time),deadline1[np.argmin(time)]))
            size1=size1-min(time)*rate
            makespan+=min(time)
            if makespan<=deadline1[np.argmin(time)]:
                print('feasible')
                feasible=1
            else:
                print('infeasible(makespan>deadline): %8.3f > %8.3f'%(makespan,deadline1[np.argmin(time)]))
                feasible=0
                break
            print('updating size:',size1)
            print('makespan:%8.3f'%makespan)
        if feasible==1:
            print('***problem feasible with makespan:%8.3f***'%makespan)
            sheet.loc[len(sheet)] = [ins,t,alpha,makespan]

        else:
            print('***problem INFEASIBLE***')
            sheet.loc[len(sheet)] = [ins,t,alpha,float('inf')]
```
