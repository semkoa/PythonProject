```python
from gurobipy import *
import numpy as np
import math
```

```python
units=[2]
nodes=['one', 'two', 'three', 'four', 'five', 'six']; 
arcs=tuplelist([('one','two'),('one' ,'three'),('two', 'one'),('two','five'),
      ('three', 'one'),( 'three', 'four'),('three', 'five' ),('four','two'),
      ('four','three'),('four','six'),('five','two'),('five','three'),('five','six'),('six','four'),('six','five')])
c={}
for arc in arcs:
    c[arc]=10
    
NoFlows=5
orig=['four','six','one','four','three']
dest =['one','five','five','two','four']
size=[66,58,94,31,1]
VV=[20.000000, 20.000000, 20.000000,30.000000,30.000000]
deadline=[3.300000,2.900000,4.700000,1.200000,0.033333]
MinAlpha=1.684570
lowerbound={0:5.969,30:5.4}
ExistPathFlag=0;
```

```python
def fun():
    # Create optimization model
    m = Model('netflow')
    # Create variables
    r= m.addVars(NoFlows, name="r",lb=0)
    y = m.addVars(NoFlows,arcs, name="y",lb=0)
    z = m.addVars(NoFlows,arcs,units, name="z",lb=0,vtype=GRB.INTEGER)
    #objective function
    m.setObjective(sum((1/deadline[f]**2)*r[f] for f in range(NoFlows)), GRB.MAXIMIZE)
    #create constraints
    capacity1=m.addConstrs(
        (sum(u*z[f,i,j,u] for u in units)>=y[f,i,j] for f in range(NoFlows) for i,j in arcs ), "capacity1")

    capacity2=m.addConstrs(
        (sum(u*z[f,i,j,u] for u in units for f in range(NoFlows))<=c[i,j] for i,j in arcs ), "capacity2")

    balance=m.addConstrs(
        (y.sum(f,'*',j)-y.sum(f,j,'*')==-r[f] for f in range(NoFlows) for j in [orig[f]] ), "balance1")
    balance2=m.addConstrs(
        (y.sum(f,'*',j)-y.sum(f,j,'*')==r[f] for f in range(NoFlows) for j in [dest[f]] ), "balance2")

    balance3=m.addConstrs(
        (y.sum(f,'*',j)-y.sum(f,j,'*')==0 
         for f in range(NoFlows) for j in list(set(nodes)-set().union([orig[f]],[dest[f]])) ), "balance3")
    m.optimize()
    m.write('netflow.lp')
    print('------------------------------------------------------------------------------------')
```

```python
fun()
```
